```python
import speech_recognition as sr
from pydub import AudioSegment
from pydub.silence import split_on_silence
from tqdm.notebook import tqdm
import os
```

```python
r = sr.Recognizer()

# a function that splits the audio file into chunks and applies speech recognition
def get_podcast_transcription(path):

    """
    Splitting the large audio file into chunks
    and apply speech recognition on each of these chunks
    """

    sound = AudioSegment.from_wav(path)  
    chunks = tqdm(split_on_silence(sound,
                              min_silence_len=400,
                              silence_thresh=sound.dBFS-14,
                              keep_silence=400
                             ))

    folder_name = "audio-chunks"
    if not os.path.isdir(folder_name):
        os.mkdir(folder_name)
    
    transcript = ""
 
    for i, chunk in enumerate(chunks, start=1):
        # export audio chunk and save it in the `folder_name` directory.
        chunk_filename = os.path.join(folder_name, f"chunk{i}.wav")
        chunk.export(chunk_filename, format="wav")
        # recognize the chunk
        with sr.AudioFile(chunk_filename) as source:
            audio_listened = r.record(source)
            # try converting it to text
            try:
                text = r.recognize_google(audio_listened, language="ru_RU")
            except sr.UnknownValueError as e:
                print("Error:", str(e))
            else:
                text = f"{text.capitalize()}. "
                print(chunk_filename, ":", text)
                transcript += text
    # return the text for all chunks detected
    return transcript
```

```python
import os

directory = os.fsencode("C:/Users/Daria/ipynbs/wavs_error")
    
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    try:
        transcript = get_podcast_transcription("C:/Users/Daria/ipynbs/wavs_error/" + filename)
        with open(f'C:/Users/Daria/ipynbs/transcripts/{filename}.txt', 'a') as f:
            f.write(transcript)
    except:
        print(f"RequestError of {filename} transcription")
        continue
```
